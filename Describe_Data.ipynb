```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Read data from CSV file
df = pd.read_csv('earthquakes.csv')
```

```python
df.drop(['id', 'date', 'title', 'url', 'detailUrl', 'ids', 'sources',
           'types', 'net', 'code', 'geometryType', 'status',
           'postcode', 'what3words', 'locationDetails'], axis=1, inplace=True)
```

```python
# Replace Null value of column 'alert' to unknown 
df.fillna({'alert': 'unknown'}, inplace=True)

# NA-value columns will be filled by 'unknown' value
na_column = df.columns[df.isna().any()].tolist()

for column in na_column:
    df.fillna({column: 'unknown'}, inplace=True)
```

```python
# Count value in label collumn
label_counts = df['alert'].value_counts()

# Draw figure
plt.figure(figsize=(10, 6))
label_counts.plot(kind='bar', color='skyblue')
plt.title('Biểu đồ tương quan của alert')
plt.xlabel('alert')
plt.ylabel('Số lượng')
plt.xticks(rotation=0)
plt.show()
```
